```python
# eda_and_preprocessing.py

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("training data.csv")


print("First five rows of the dataset:")
print(df.head(), "\n")


print("Class distribution (condition):")
print(df["condition"].value_counts(), "\n")


numeric_cols = [
    "corrosion rate (Mpy)", "Diameter Size (inch)", "Probe Type",
    "Line Length (KM)", "Pigging Frequency (Days)",
    "Surfactant Dosing (Litres)"]
print("Summary statistics (numeric features):")
print(df[numeric_cols].describe(), "\n")


for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=20, kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.savefig(f"hist_{col}.png")
    plt.close()


plt.figure(figsize=(6,4))
sns.boxplot(x="condition", y="corrosion rate (Mpy)", data=df)
plt.title("corrosion rate (Mpy) by Condition")
plt.tight_layout()
plt.savefig("box_corrosion_by_condition.png")
plt.close()


categorical_cols = ["Probe Type"]
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f"Count of {col}")
    plt.tight_layout()
    plt.savefig(f"count_{col}.png")
    plt.close()
```
